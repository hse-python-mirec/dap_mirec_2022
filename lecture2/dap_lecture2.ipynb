```python
df['Normal Humidity'] = df['Humidity'].apply(lambda x:
                    1 if x in range(40, 61) else 0)
```

```python
df.head(3)
```

```python
def get_temp_cat(temp):
    if temp < 0:
        return 'Freezing'
    elif temp < 15:
        return 'Chilly'
    elif temp < 26:
        return 'Nice'
    elif temp >= 26:
        return 'Hot'
    else:
        return temp
```

```python
df['Temperature Category'] = df['Temperature'].apply(
                                get_temp_cat)
```

```python
df.iloc[100:234]
```

```python
df['Temperature Category'] = 
pd.Categorical(df['Temperature Category'])
```
